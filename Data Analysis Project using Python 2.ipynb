##1) Instruction (For Data Cleaning)

*  Find all Null Values in the dataset. If there is any Null Value in the Column,then fill it with the mean of that Column.


```python
car.isnull().sum()
```

```python
car_numeric_cols = car.select_dtypes(include=['float64', 'int64'])
car[car_numeric_cols.columns] = car_numeric_cols.fillna(car_numeric_cols.mean())
```

```python
for col in car.select_dtypes(include='object'):
    car[col].fillna(car[col].mode()[0], inplace=True)
```
